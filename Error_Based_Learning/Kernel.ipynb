```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
data = {'x':[1,-1,-1,1,2,-2,-2,2,4,-4,-4,4,5,-5,-5,5],
      'y':[1,1,-1,-1,2,2,-2,-2,4,4,-4,-4,5,5,-5,-5],
      'class':['a','a','a','a','a','a','a','a','b','b','b','b','b','b','b','b']}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
sns.scatterplot(data=df, x='x', y='y', hue='class')
plt.show()
```

```python
# transform into higher dimension
# performing the heavy lifting of linear separability
df['z'] = df['x']**2 + df['y']**2

df = df[['x','y','z','class']]
```

```python
df
```

```python
fig = px.scatter_3d(df, x='x', y='y', z='z',
              color='class')


## separating hyperplane
light_yellow = [[0, '#FFDB58'], [1, '#FFDB58']]

zero_pt = pd.Series([])
z = zero_pt.append(df['z'], ignore_index = True).reset_index(drop = True)
y = zero_pt.append(df['y'], ignore_index = True).reset_index(drop = True)
x = zero_pt.append(df['x'], ignore_index = True).reset_index(drop = True)


length_data = len(z)
z_plane_pos = 20*np.ones((length_data,length_data))

fig.add_trace(go.Surface(x=x, y=y, z=z_plane_pos, colorscale=light_yellow,  showscale=False))

fig.show()
```
